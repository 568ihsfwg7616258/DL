```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن داده‌ها
Data = pd.read_csv('./employerMoney.txt', sep=';', dtype=str)

# تنظیم نام ستون‌ها
Data.columns = ['Company', 'Rep', 'Dem', 'Other', 'Total']

# نمایش اطلاعات اولیه داده‌ها
print("Data shape:", Data.shape)
print(Data.head())

# چاپ نام شرکت‌ها
print(Data['Company'])

# انتخاب ردیف‌های خاص برای ترسیم
s = list(range(10))  # انتخاب 10 ردیف اول

# اطمینان از اینکه بازه انتخابی در محدوده داده‌ها قرار دارد
if max(s) < Data.shape[0]:
    D = Data.iloc[s].copy()
    
    # تبدیل ستون‌های Rep و Dem به عددی و بررسی مقادیر نامعتبر
    D['Rep'] = pd.to_numeric(D['Rep'], errors='coerce')
    D['Dem'] = pd.to_numeric(D['Dem'], errors='coerce')
    
    # حذف ردیف‌هایی که مقادیر نامعتبر دارند
    D = D.dropna(subset=['Rep', 'Dem'])
    
    # مرتب‌سازی داده‌ها بر اساس جمع کل ستون‌های Rep و Dem
    D = D.sort_values(by=['Rep', 'Dem'])
    
    # مقیاس‌دهی مقادیر
    rep = D['Rep'] / 10**5
    dem = D['Dem'] / 10**5
    
    # یافتن مقدار حداکثر برای ترسیم
    mx = (rep + dem).max()
    
    # استخراج نام شرکت‌ها
    names = D['Company']
    n = len(rep)
    
    # ترسیم داده‌ها
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # ترسیم خطوط
    for i in range(n):
        ax.plot([0, rep.iloc[i]], [i + 1, i + 1], color='red', linewidth=3)
        ax.plot([rep.iloc[i], rep.iloc[i] + dem.iloc[i]], [i + 1, i + 1], color='blue', linewidth=3)
    
    # اضافه کردن برچسب‌های محور y
    ax.set_yticks(range(1, n + 1))
    ax.set_yticklabels(names, fontsize=8)
    ax.set_xlabel("Dollars - 100,000's", fontsize=10)
    ax.set_ylim(0, n + 1)
    
    plt.tight_layout()
    plt.show()
else:
    print("Selected row indices are out of bounds.")
```

```python
simDict = {}
for commA, commB in pairs:
    A = contributorDict[commA] # Set of contributors to commA.
    nameA = nameDict[commA]
    B = contributorDict[commB]
    nameB = nameDict[commB]

    nIntersection = len(A.intersection(B))
    jAB = nIntersection/len(A.union(B))
    pAGivenB = nIntersection/len(B)
    pBGivenA = nIntersection/len(A)
    simDict[(nameA, nameB)] = (jAB, pAGivenB, pBGivenA)
```
